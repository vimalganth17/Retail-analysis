# Key performance indicator


```python
total_sales = df['Sales'].sum()
print(f'Total Sales: ${total_sales:.2f}')
```

```python
average_sales_per_order = df.groupby('Order ID')['Sales'].sum().mean()
print(f'Average Sales per Order: ${average_sales_per_order:.2f}')
```

```python
total_profit = df['Profit'].sum()
print(f'Total Profit: ${total_profit:.2f}')
```

```python
average_profit_margin = (df['Profit'] / df['Sales']).mean()
print(f'Average Profit Margin: {average_profit_margin * 100:.2f}%')
```

```python
total_orders = df['Order ID'].nunique()
print(f'Total Orders: {total_orders}')
```

```python
top_selling_products = df.groupby('Product Name')['Quantity'].sum().nlargest(5)
print('Top Selling Products:')
print(top_selling_products)
```

```python
top_selling_products = df.groupby('Product Name')['Quantity'].sum().nlargest(5)

plt.figure(figsize=(12, 6))
top_selling_products.sort_values().plot(kind='bar')
```

```python

product_sales = df.groupby('Product Name')['Sales'].sum()

top_n = 10
top_products = product_sales.nlargest(top_n)


top_products.sort_values().plot(kind='barh')

plt.show()
```

```python
average_discount_rate = df['Discount'].mean()
print(f'Average Discount Rate: {average_discount_rate * 100:.2f}%')
```

```python
average_discount_rate = df['Discount'].mean()

plt.figure(figsize=(8, 8))
labels = ['Average Discount', 'Remaining']
sizes = [average_discount_rate, 1 - average_discount_rate]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Average Discount Rate Visualization')
plt.show()
```

```python
customer_acquisition_rate = df['Customer ID'].nunique() / total_orders
print(f'Customer Acquisition Rate: {customer_acquisition_rate:.2f} customers per order')
```

```python
market_sales_profit = df.groupby('Country')['Sales', 'Profit'].agg(['sum'])

print('Sales and Profit Per Market:')
print(market_sales_profit)
```

```python
market_sales_profit = df.groupby('Country')['Sales', 'Profit'].agg(['sum'])

market_sales_profit.plot(kind='bar', color=['skyblue', 'lightgreen'])

plt.legend(['Sales', 'Profit'])
plt.show()
```

```python

market_sales_profit = df.groupby('Country')['Sales', 'Profit'].agg(['sum']).reset_index()

fig, ax = plt.subplots(figsize=(12, 6))

sns.barplot(x='Country', y=('Sales', 'sum'), data=market_sales_profit, ax=ax, label='Sales')

sns.barplot(x='Country', y=('Profit', 'sum'), data=market_sales_profit, ax=ax, label='Profit')


plt.show()
```

# Which Month does the company record more sales?

```python

df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract month from 'Order Date'
df['Order Month'] = df['Order Date'].dt.month

# Group by month and calculate total sales
monthly_sales = df.groupby('Order Month')['Sales'].sum()
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='bar', color='skyblue')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Total Sales by Month')

plt.show()
```

```python
market_category_sales = df.groupby(['Country', 'Category'])['Sales'].sum().reset_index()

# Find the dominant product category in each market
dominant_categories = market_category_sales.loc[market_category_sales.groupby('Country')['Sales'].idxmax()]
print(dominant_categories)
```

# The city with the most orders for a given product category

```python


# Specify the product category of interest
desired_category = 'Office Supplies'
category_orders = df[df['Category'] == desired_category]
city_order_counts = category_orders.groupby('City')['Order ID'].nunique()

# Find the city with the most orders
most_orders_city = city_order_counts.idxmax()
most_orders_count = city_order_counts.max()

print(f"For the product category '{desired_category}', the city with the most orders is {most_orders_city} with {most_orders_count} orders.")
```
